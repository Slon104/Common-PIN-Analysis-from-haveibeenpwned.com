```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Read and process the data
data = {}
with open('HIBP_4_digit_wordlist_heat.txt', 'r') as file:
    for line in file:
        key, value = line.strip().split(' : ')
        data[key] = int(value)

# Create a DataFrame
df = pd.DataFrame(list(data.items()), columns=['Code', 'Value'])
df['X'] = df['Code'].str[:2].astype(int)
df['Y'] = df['Code'].str[2:].astype(int)

# Step 2: Create a 2D array for the heatmap
heatmap = np.zeros((100, 100))
for _, row in df.iterrows():
    heatmap[row['Y'], row['X']] = row['Value']

# Step 3: Apply a power transformation and normalize the data
power = 0.3  # Choose a power less than 1 to enhance lower values
heatmap_power = np.power(heatmap, power)

min_value = heatmap_power.min()
max_value = heatmap_power.max()

if min_value == max_value:
    heatmap_normalized = np.zeros_like(heatmap_power)
else:
    heatmap_normalized = (heatmap_power - min_value) / (max_value - min_value)

# Round the normalized values to six decimal places
heatmap_normalized = np.round(heatmap_normalized, 6)

# Step 4: Create the heatmap
plt.figure(figsize=(10, 10))
plt.imshow(heatmap_normalized, cmap='hot', interpolation='nearest', origin='lower')
plt.colorbar(label='Normalized Value (Power Transformed)')
plt.title('Heatmap of 4-digit Codes (Power Transformed, p=0.3)')
plt.xlabel('First Two Digits')
plt.ylabel('Last Two Digits')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Step 1: Read and process the data
data = {}
with open('all_6digit_pins_with_counts.txt', 'r') as file:
    for line in file:
        key, value = line.strip().split(' : ')
        data[key] = int(value)

# Create a DataFrame
df = pd.DataFrame(list(data.items()), columns=['Code', 'Value'])
df['X'] = df['Code'].str[:2].astype(int)
df['Y'] = df['Code'].str[2:4].astype(int)
df['Z'] = df['Code'].str[4:].astype(int)

# Step 2: Apply a power transformation to the values
power = 0.2  # Choose a power less than 1 to enhance lower values
df['Value_transformed'] = np.power(df['Value'], power)

# Step 3: Normalize the transformed values
min_value = df['Value_transformed'].min()
max_value = df['Value_transformed'].max()
if min_value == max_value:
    df['Value_normalized'] = 0
else:
    df['Value_normalized'] = (df['Value_transformed'] - min_value) / (max_value - min_value)

# Round the normalized values to six decimal places
df['Value_normalized'] = df['Value_normalized'].round(6)

# Step 4: Create the 3D scatter plot
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(df['X'], df['Y'], df['Z'], c=df['Value_normalized'], cmap='hot', s=5)

ax.set_xlabel('First Two Digits')
ax.set_ylabel('Middle Two Digits')
ax.set_zlabel('Last Two Digits')
ax.set_title('3D Visualization of 6-digit Codes [ABCDEF] (Power Transformed, p=0.2)')

# Add a color bar
cbar = plt.colorbar(scatter)
cbar.set_label('Normalized Value (Power Transformed)')

plt.show()
```
